```python
df=df.drop(columns=["id","date","zipcode"],axis=1)
df.head()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 17 , 15 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
```

The strongest correlation that Price has is the one with the surface of the living("sqft living" = 0.7).It is also well correlated with : the "Grade" given to the construction (0.67), the surface of the surface above ground floor("sqft above"=0.61) and the square footage of interior living space for the nearest 15 neighbours ("sqft living15"=0.59).
We note also a a relatively moderate correlation with the number of "bathrooms" (0.53).All of the other features do not have a big influence on the house pricings . So, we should drop them.

```python
df=df.drop(columns=["sqft_lot15","long","lat","yr_renovated","yr_built","condition","waterfront","floors",
                   "sqft_lot","bedrooms","view"],axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
S=StandardScaler()
df1=df.copy()
df1=S.fit_transform(df1)
box_plot=sns.boxplot(data=df1)
box_plot.set(xlabel=["Price","Bathrooms","sqft_living","Grade","sqft_above","sqft_living15"])
```

```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 

from sklearn.metrics import mean_squared_error
from sklearn import metrics
x= df["sqft_living"].values[:,np.newaxis]
y= df["price"].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  
lg=LinearRegression()
lg.fit(x_train, y_train)
predicted=lg.predict(x_test)
```

```python
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("Square Footage of Living")
plt.xlabel("Price")
plt.plot(x_train,lg.predict(x_train)) #testing our modelâ€™s performance,color="k")
plt.show()
```

```python

print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

=> Linear regression isn't the best model to express the relation between the price and the sqft_living since the MSE is too high and the R squared is too small.

```python
x= df[["sqft_living","grade","sqft_above","sqft_living15","sqft_basement","bathrooms"]]
y= df["price"].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  
lg=LinearRegression()
lg.fit(x_train, y_train)
predicted=lg.predict(x_test)
```

```python
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

=> even though the Multilinear regression's MSE is lower than the Linear model's one, and the R squared is bigger than the one obtained with the Linear regression model.
Multilinear regression can't be used to predict the houses price. 

```python
poly = PolynomialFeatures(degree =2) 
x= df[["sqft_living","grade","sqft_above","sqft_living15","bathrooms"]]
y= df["price"].values
x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)
print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
x= df["sqft_living"].values.reshape(-1,1)
y= df["price"].values
poly = PolynomialFeatures(degree = 2) 
x_poly = poly.fit_transform(x) 
poly.fit(x_poly, y) 
lg=LinearRegression()
lg.fit(x_poly, y) 

plt.scatter(x, y, color="g")
plt.title("Linear regression")
plt.ylabel("Price")
plt.xlabel("sqft_living")
plt.plot(x, lg.predict(poly.fit_transform(x)), color="r") 

```

The polynomial regression is the best model to choose for this dataset among the models that we've seen through our course since it has a quite important R squared compared to other models (0.66) and a smaller MSE so less predictive errors.
